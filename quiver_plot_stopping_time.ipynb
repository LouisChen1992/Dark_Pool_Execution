```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x, y, lam=0.1, alpha=np.log(10)):
    return 1 - np.exp(- lam * alpha * y / x)
```

```python
lam = 1.0
alpha = np.log(10)
R = 2

delta=0.25
lam_delta = - lam * alpha / np.log(1 - delta)
x_0 = delta
y_0 = delta / lam_delta
```

```python
x = np.arange(0.0, 2.0, 0.2)
y = np.arange(0.0, 2.0, 0.2)
X, Y = np.meshgrid(x, y)
```

```python
U = g(X, Y, lam=lam, alpha=alpha) - X
V = g(X, Y, lam=lam, alpha=alpha) / lam - Y
```

```python
plt.quiver(X, Y, U, V, units='width')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.show()
# plt.savefig('quiver.pdf')
```

```python
Y1 = np.arange(y_0, np.sqrt(R**2 - delta**2) / lam, 0.2)
X1 = np.ones_like(Y1) * x_0

X2 = np.arange(x_0, lam_delta * R / np.sqrt(lam_delta**2 + lam**2), 0.2)
Y2 = np.sqrt(R**2 - X2**2) / lam

X3 = np.array([lam_delta * R / np.sqrt(lam_delta**2 + lam**2)])
Y3 = X3 / lam_delta

X4 = X2[::-1]
Y4 = X4 / lam_delta

X5 = np.arange(x_0, R / np.sqrt(2), 0.2)
Y5 = X5 / lam

X6 = np.array([R / np.sqrt(2)])
Y6 = X6 / lam
```

```python
XX = np.concatenate([X1, X2, X3, X4, X5, X6])
YY = np.concatenate([Y1, Y2, Y3, Y4, Y5, Y6])
UU = g(XX, YY, lam=lam, alpha=alpha) - XX
VV = g(XX, YY, lam=lam, alpha=alpha) / lam - YY
```

```python
plt.plot(XX, YY, '-.')
plt.quiver(XX, YY, UU, VV, units='width')
plt.xlim(0.0, 2.1)
plt.ylim(0.0, 2.1)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.text(0.5, 1.0, r'$K_1$', fontsize=30)
plt.text(1.0, 0.4, r'$K_2$', fontsize=30)
plt.show()
# plt.savefig('quiver_K1_K2.pdf')
```
