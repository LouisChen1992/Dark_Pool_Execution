```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g_delta(x, y, lam=0.1, alpha=np.log(10), delta=0.5):
    return 1 - np.exp(- lam * alpha * y / (x + delta))

def f_delta(x, lam=0.1, alpha=np.log(10), delta=0.5):
    return g_delta(x, x / lam, lam, alpha, delta)
```

```python
x = np.arange(0.0, 1.5, 0.1)
y = np.arange(0.0, 1.5, 0.1)
X, Y = np.meshgrid(x, y)
alpha = np.log(10)
lam = 1.0
delta=0.5
R = np.sqrt(2)
eta = 3 * lam
```

```python
plt.axhline(y=1-np.exp(-alpha), xmin=0, xmax=x[-1], color='black')
plt.plot(x, f_delta(x, lam=lam, alpha=alpha, delta=delta))
plt.plot(x, x)
plt.xlabel('x')
plt.show()
# plt.savefig('equilibrium_x.pdf')
```

```python
U = g_delta(X, Y, lam=lam, alpha=alpha, delta=delta) - X
V = g_delta(X, Y, lam=lam, alpha=alpha, delta=delta) / lam - Y
```

```python
plt.quiver(X, Y, U, V, units='width')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.show()
# plt.savefig('quiver.pdf')
```

```python
Y1 = np.arange(0,R / lam, 0.1)
X1 = np.zeros_like(Y1)

X2 = np.arange(0, R / np.sqrt(1 + lam**2 / eta**2), 0.1)
Y2 = X2 / eta

X3 = np.arange(0, R / np.sqrt(1 + lam**2 / eta**2), 0.1)[::-1]
Y3 = np.sqrt(R**2 - X3**2) / lam

X4 = np.array([R / np.sqrt(1 + lam**2 / eta**2)])
Y4 = X4 / eta

X5 = np.arange(0, R / np.sqrt(2), 0.1)[::-1]
Y5 = X5 / lam

X6 = np.arange(0, R, 0.1)
Y6 = np.zeros_like(X6)

X7 = np.arange(0, R, 0.1)
Y7 = np.sqrt(R**2 - X7**2) / lam

X8 = np.array([R / np.sqrt(2)])
Y8 = X8 / lam

X9 = np.array([R])
Y9 = np.array([0])
```

```python
XX = np.concatenate([X1, X2, X4, X3])
YY = np.concatenate([Y1, Y2, Y4, Y3])
UU = g_delta(XX, YY, lam=lam, alpha=alpha, delta=delta) - XX
VV = g_delta(XX, YY, lam=lam, alpha=alpha, delta=delta) / lam - YY
```

```python
plt.plot(XX, YY, '-.')
plt.quiver(XX, YY, UU, VV, units='width')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.text(0.4, 0.7, r'$K$', fontsize=30)
plt.show()
# plt.savefig('quiver_K.pdf')
```

```python
XX = np.concatenate([X8, X5, X1, X7, X9, X6])
YY = np.concatenate([Y8, Y5, Y1, Y7, Y9, Y6])
UU = g_delta(XX, YY, lam=lam, alpha=alpha, delta=delta) - XX
VV = g_delta(XX, YY, lam=lam, alpha=alpha, delta=delta) / lam - YY
```

```python
plt.plot(XX, YY, '-.')
plt.quiver(XX, YY, UU, VV, units='width')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.text(0.3, 0.7, r'$K_1$', fontsize=30)
plt.text(0.7, 0.3, r'$K_2$', fontsize=30)
plt.show()
# plt.savefig('quiver_K1_K2.pdf')
```
